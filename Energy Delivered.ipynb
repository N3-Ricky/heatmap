### Add library

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# Energy Delivered - Detail

```python
EDD = pd.read_excel('Data Cleaning.xlsx', sheet_name = 'Energy Delivered - Detail')
#print(EDD)
##EDD.head()
```

```python
EDDi = EDD[['State','Year','GWh']]
#print(EDDi)
```

```python
EDDx = EDDi.groupby(['State','Year']).mean()
#print(EDDx)
#EDDx.to_excel('Heatmap.xlsx', sheet_name = 'EDDx')
```

```python
EDDx = EDDx.reset_index()
#print(x)
```

```python
#EDDx.pivot('State','Year','GWh').head
```

```python
plt.figure(figsize=(18,9))
pivot_table = EDDx.pivot('State','Year','GWh')
plt.xlabel('State',size = 20)
plt.ylabel('Year',size = 20)
plt.title('Energy Delivered - Detail',size = 20)
sns.heatmap(pivot_table, annot=True, fmt=".1f", linewidths=0.5 , square = True, cmap = 'Blues')
```

# Average Outage Duration

```python
AOD = pd.read_excel('Data Cleaning.xlsx', sheet_name = 'Avg Outage Duration (Detailed)')
#print(NU)
AODi = AOD[['State','Year','Outage Duration in minutes']]
#print(i)
AODx = AODi.groupby(['State','Year']).mean()
#NUx.head()
#AODx.to_excel('Copy of Data Cleaning.xlsx', sheet_name = 'AODx')
AODx = AODx.reset_index()
#print(NUx)
AODx.pivot('State','Year','Outage Duration in minutes').head

plt.figure(figsize=(18,9))
pivot_table = AODx.pivot('State','Year','Outage Duration in minutes')
plt.xlabel('State',size = 20)
plt.ylabel('Year',size = 20)
plt.title('Outage Duration in minutes',size = 20)
sns.heatmap(pivot_table, annot=True, fmt=".1f", linewidths=0.5 , square = True, cmap = 'Blues')
```
